```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
```

```python
img = im.imread(r"C:\Python36\images\image_data\image_data\np2.jpg")
plt.imshow(img)
plt.show()
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img2,cmap='gray')
plt.show()
```

```python
print(img.shape)
print(img2.shape)
```

```python
# Morphological analysis
# erosion 
# Dilation
```

```python
fac = np.ones(5)
```

```python
# erosion
img3 = cv2.erode(img2,fac)
plt.imshow(img3,cmap='gray')
plt.show()
```

```python
# Dilation
img4 = cv2.dilate(img3,fac)
plt.imshow(img4,cmap='gray')
plt.show()
```

```python
# Edge Detection
img = cv2.imread(r"C:\Python36\images\image_data\image_data\car3.png",0)
plt.imshow(img,cmap="gray")
plt.show()
```

```python
edges = cv2.Canny(img,128,200)
plt.imshow(edges,cmap="gray")
plt.show()
```
